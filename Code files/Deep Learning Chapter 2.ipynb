## Let's see the simple example Neural network: 



```python
# make a function that takes input data multiplies it with weight and provides the result as a output prediction
import numpy as np
weight = 0.1
def neural_network(inputs, weight):
    prediction = inputs*weight
    return prediction

```

```python
number_of_toes = [8.5, 9.5, 10, 9]
input = number_of_toes[0]
pred = neural_network(input,weight)
print(pred)
```
